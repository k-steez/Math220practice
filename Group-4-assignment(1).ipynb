```python

def unityfunction(n): 
    rootlist =[]
    for k in range(n): 
        z = math.cos((2*k*math.pi)/n) + 1j * math.sin((2*k*math.pi)/n)
        rootlist.append(z)
    return rootlist
```

```python
rootlist = unityfunction(100)
```

```python
def equalspace(lst,n):
    for i in range(n-1):
        a = lst[1] - lst[0]
        if lst[i+1] - lst[i] == a:
            print('equally spaced')
        else:
            print('unequally spaced')
        
        
```

```python
def plot(lst):
    x = []
    y = []
    for k in range(len(lst)):
        x.append(lst[k].real)
        y.append(lst[k].imag)
    return x,y
```

```python
x,y = plot(rootlist)
```

```python
plt.plot(x,y)
plt.gra().set_aspect('equal')
```
